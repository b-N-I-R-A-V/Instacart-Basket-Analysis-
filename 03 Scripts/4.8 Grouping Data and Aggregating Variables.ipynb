### Creating a new column avg_price which stores average price paid by a user_id

```python
# Fixing outliers before exporting
ords_prods_merge.loc[ords_prods_merge['prices'] >100, 'prices'] = np.nan
```

```python
ords_prods_merge['avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#Checking the column
ords_prods_merge.head()
```

```python
# Creating a flag column to flag customer as low spender or high spender
ords_prods_merge.loc[ords_prods_merge['avg_price'] < 10, 'spending_flag'] = 'Low Spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_price'] >= 10, 'spending_flag'] = 'High Spender'
```

### Determining Frequent and Non-frequent Customers

```python
# Calculating median of days_since_prior_order
ords_prods_merge['median_days_bw_orders'] = ords_prods_merge.groupby('user_id')['days_since_prior_order'].transform(np.median)
```

If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# Creating flags for frequent, non-frequent, and regular customer
ords_prods_merge.loc[ords_prods_merge['median_days_bw_orders'] > 20, 'order_frequency'] = 'Non-frequent customer' 
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_bw_orders'] > 10) & 
                     (ords_prods_merge['median_days_bw_orders'] <= 20), 'order_frequency'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_bw_orders'] <= 10, 'order_frequency'] = 'Frequent customer'
```
